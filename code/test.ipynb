```python
import numpy as np
import matplotlib.pyplot as plt

Ns = [50, 100, 500, 2000]
xs = np.arange(1, 1000, 1)
for N in Ns:
    ys = (1/(2*N))*np.exp(-xs/(2*N))
    plt.plot(xs, ys, label=f'N={N}')

plt.legend(loc='upper right')
plt.show()
```
