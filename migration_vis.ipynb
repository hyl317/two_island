```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

bins = np.arange(5, 20, 0.2)
midpoint = (bins[1:] + bins[:-1])/2

nomig = pickle.load(open('/mnt/archgen/users/yilei/IBD/two_island_final/inference/IBD_pickle/T20_N500/ibd_batch1.pickle', 'rb'))
print(len(nomig))
x, _ = np.histogram(nomig, bins=bins)
x = x/100
i = np.where(x != 0)[0]
plt.scatter(midpoint[i], x[i], s=7.0, label='no migration')

mig = pickle.load(open('/mnt/archgen/users/yilei/IBD/two_island_final/Inference_migration/IBD_pickle/M4_N500/ibd_batch1.pickle', 'rb'))
print(len(mig))
x, _ = np.histogram(mig, bins=bins)
x = x/100
i = np.where(x != 0)[0]
plt.scatter(midpoint[i], x[i], s=7.0, label='migration rate: 0.02')


plt.yscale('log')
plt.legend(loc='upper right', fontsize='large')
plt.xlabel('IBD Segment Length (cM)')
plt.ylabel('Number of Segments per Pair of Samples')
plt.savefig('mig_comparison.pdf')
plt.savefig('mig_comparison.png', dpi=300)
```
